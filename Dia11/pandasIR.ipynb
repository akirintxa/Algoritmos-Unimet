EJ 5. Indique quienes fueron los 5 jugadores más efectivos, para ello debe sumar el total de goles y asistencias de un solo jugador y guardarlo en una nueva columna.

```python
df['efectividad'] = df['goles'] + df['asistencias']
df
```

```python
agrupacion_nombre_efectividad = df.groupby(['nombre_jugador'])[['efectividad']].sum()
agrupacion_nombre_efectividad
```

```python
top_efectivos = agrupacion_nombre_efectividad.sort_values("efectividad",ascending=False).head()
top_efectivos
```
