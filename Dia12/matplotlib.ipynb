## Pyplot

El metodo **.subplots()** creara dos objetos:

- Una **Figure** *fig* (figura) que es un contenedor de todo lo que se verá en el gráfico.
- Los **Axes** *ax* (ejes) que contendran la información a graficar.

````python
# Crea un grafico vacio
fig, ax = plt.subplots()

# Mostramos el grafico con .show()
plt.show()
````

## Grafico de linea

### Añadir datos

Para poder visualizar los datos, tendremos que añadirlos a los ejes. Para ello, podemos usar los **metodos de ploteo** de los objetos **Axes**.

El metodo más básico de todos es **.plot()**, el cual crea un **gráfico lineal** y que requiere los *datos del eje X y Y como argumentos*.

````python
# Cargamos el csv
df = pd.read_csv('futbol_estadisticas.csv', index_col='nombre_jugador')
````

````python
# Filtramos la data para mostrar solo los datos del jugador 2097
# Recordemos que loc nos permite seleccionar filas especificas
filas_messi = df.loc['Lionel Messi']

filas_messi
````

````python
# Creamos un grafico vacio
fig, ax = plt.subplots()

# Plot -> ax.plot(X, Y)
ax.plot(filas_messi["año"], filas_messi["goles"])

# Mostramos el grafico
plt.show()
````

### Guardar graficos

Tambien podemos guardar los graficos generados con el **metodo .savefig()**.

````python
# Creamos un grafico vacio
fig, ax = plt.subplots()

# Plot -> ax.plot(X, Y)
ax.plot(filas_messi["año"], filas_messi["goles"])

# Guardamos el grafico
plt.savefig('messi.jpg')

# Mostramos el grafico
plt.show()
````

### Mas de una grafica en una figura

Puede haber más de una sola gráfica para una misma figura, solo tenemos que usar **.plot()** más de una vez y se generaran distintas gráficas para un mismo eje.

````python
# Filtramos la data para mostrar solo los datos del jugador 2371
# Recordemos que loc nos permite seleccionar filas especificas
filas_cristiano = df.loc['Cristiano Ronaldo']

filas_cristiano
````

````python
# Creamos un grafico vacio
fig, ax = plt.subplots()

# Plot -> ax.plot(X, Y)
ax.plot(filas_messi["año"], filas_messi["goles"])
ax.plot(filas_cristiano["año"], filas_cristiano["goles"])

# Guardamos el grafico
plt.savefig('messi_vs_cristiano.jpg')

# Mostramos el grafico
plt.show()
````

### Personalizacion de un plot

El gráfico anterior está bien, pero no es muy claro. Resulta difícil comprender que linea pertenece cada jugador y qué representa cada eje. 

Por esto, existen una **serie de argumentos** que nos permitirán personalizar cómo se ven los datos del plot:

- **label = 'nombre'** → Coloca una **etiqueta** al plot para saber a qué corresponde cada color *(útil cuando hay varias gráficas en un solo plot)*.
- **marker = 'o'** → Añade **marcadores** al plot y se le puede indicar el tipo de marcador preferido.
  - [Opciones de marcadores](https://matplotlib.org/stable/api/markers_api.html)
- **linestyle = '--'** → Cambia el **tipo de línea**. 
  - [Opciones de tipo de linea](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)
  - Si se coloca **'None'**, no habrá líneas y quedarán solo los marcadores.
- **color = 'b'** → Cambia el color de los datos. 
  - [Colores](https://matplotlib.org/stable/tutorials/colors/colors.html)

````python
# Creamos un grafico vacio
fig, ax = plt.subplots()

# Plot -> ax.plot(X, Y)
ax.plot(filas_messi["año"], # X
        filas_messi["goles"], # Y
        marker = 'o', # circulo
        color = 'b', # azul
        linestyle = '--', # dashed
        label = 'Lionel Messi') # etiqueta

ax.plot(filas_cristiano["año"], # X
        filas_cristiano["goles"], # Y
        marker = 's', # cuadrado
        color = 'r', # rojo
        linestyle = 'dotted', # dotted
        label = 'Cristiano Ronaldo') # etiqueta

# Mostramos el grafico
plt.show()
````

Ademas, el **objeto Axes** tiene métodos que nos permiten indicar de qué trata la gráfica. 

- **.set_xlabel('Nombre')** → Coloca un nombre al eje de las X.
- **.set_ylabel('Nombre')** → coloca un nombre al eje de las Y.
- **.set_title('Nombre')** → coloca un título al plot.
- **.set(xlabel='Nombre', ylabel='Nombre')** → Tambien se puede usar únicamente set y pasar como parámetros lo anterior.
- **.legend()** → Podemos mostrar las etiquetas (label) creadas con **.legend()**.

````python
# Creamos un grafico vacio
fig, ax = plt.subplots()

# Plot -> ax.plot(X, Y)
ax.plot(filas_messi["año"], # X
        filas_messi["goles"], # Y
        marker = 'o', # circulo
        color = 'b', # azul
        linestyle = '--', # dashed
        label = 'Lionel Messi') # etiqueta

ax.plot(filas_cristiano["año"], # X
        filas_cristiano["goles"], # Y
        marker = 's', # cuadrado
        color = 'r', # rojo
        linestyle = 'dotted', # dotted
        label = 'Cristiano Ronaldo') # etiqueta

# Titulo, etiquetas
ax.set(xlabel='Años', ylabel='Goles', title='Messi vs. Cristiano (2014 - 2020)')

# Leyenda
ax.legend()

# Mostramos el grafico
plt.show()
````

**ADVERTENCIA**

La personalizacion siempre se debe realizar antes de mostrar la gráfica (antes de **plt.show()**).

### Subplots

Podemos pasar toda la información que queramos y podriamos tener varios gráficos uno encima de otro, esto a veces puede ser útil.

Sin embargo, a veces esto puede generar un completo **desastre**:

````python
# Filtramos la data para mostrar solo los datos del jugador 3294
# Recordemos que loc nos permite seleccionar filas especificas
filas_cavani = df.loc['Edinson Cavani']

filas_cavani
````

````python
# Creamos un grafico vacio
fig, ax = plt.subplots()

# Plot -> ax.plot(X, Y)
ax.plot(filas_messi["año"], # X
        filas_messi["goles"], # Y
        marker = 'o', # circulo
        color = 'b', # azul
        linestyle = '--', # dashed
        label = 'Lionel Messi') # etiqueta

ax.plot(filas_messi["año"], # X
        filas_messi["goles_sin_penalti"], # Y
        marker='.', # punto
        color='b', # azul
        linestyle='dotted') # dotted

ax.plot(filas_cristiano["año"], # X
        filas_cristiano["goles"], # Y
        marker = 's', # cuadrado
        color = 'r', # rojo
        linestyle = '--', # dashed
        label = 'Cristiano Ronaldo') # etiqueta

ax.plot(filas_cristiano["año"], # X 
        filas_cristiano["goles_sin_penalti"], # Y
        marker='.', # punto
        color='r', # rojo
        linestyle='dotted') # dotted

ax.plot(filas_cavani["año"], # X
        filas_cavani["goles"], # Y
        marker='D', # diamante
        color='m', # magenta
        linestyle='--', # dashed
        label='Edinson Cavani') # etiqueta

ax.plot(filas_cavani["año"], # X 
        filas_cavani["goles_sin_penalti"], # Y
        marker='.', # punto
        color='m', # magenta
        linestyle='dotted') # dotted

# Titulo, etiquetas
ax.set(xlabel='Años', ylabel='Goles', title='Goles y no penalti goles')

# Leyenda
ax.legend()

# Mostramos el grafico
plt.show()
````

Podemos crear subplots dentro de una misma figura para poder dividir la información. 

Para eso, se usa **.subplots()**. El primer argumento es un entero que hace referencia al **número de filas** y el segundo es otro entero que indica el **número de columnas**, obteniendo asi una **matriz de plots**.

````python
# Creamos un grafico vacio con 6 subplots
fig, ax = plt.subplots(3, 2)  # matriz 3 × 2

# Mostramos el grafico
plt.show()
````

Todo está contenido dentro de la Figura y los Ejes contienen la información del plot.

Hay una **matriz de plots**, significa que se creó una **matriz de axes**.

````python
# Para conocer el tamaño de la matriz de plots
ax.shape
````

Podemos acceder a cada uno de los plots con la notación 

```python
ax[fila, columna]
```

Por ejemplo, para acceder al plot ubicado en la **segunda fila** y **primera columna**, sería:

````python
ax[1, 0]
````

En caso de tener una **única fila o columna de plots**, solo tenemos que pasar el índice correspondiente al arreglo. 

Por ejemplo, para acceder al **tercer plot**:

````python
ax[2]
````

Si queremos que todos los plots tengan la **misma escala en el eje Y**, debemos pasarle a **.subplots()** el argumento **sharey = True**. 

````python
# Creamos un grafico vacio con 6 subplots
fig, ax = plt.subplots(3, 2, sharey = True)  # matriz 3 × 2

# Mostramos el grafico
plt.show()
````

Podemos usar **figsize** para colocar el ancho y alto respectivamente (en pulgadas). Por defecto, está en **[6.4, 4.8]**.

````python
# Creamos un grafico vacio con 6 subplots
fig, ax = plt.subplots(3, 2, sharey = True, figsize = [10, 10])  # matriz 3 × 2

# Mostramos el grafico
plt.show()
````

Teniendo esto en cuenta, podemos mejorar el grafico anterior:

````python
# Creamos un grafico vacio de 3 filas y una columna que comparta la escala en y
fig, ax = plt.subplots(3, 1, sharey = True, figsize = [6, 10])

# Plot -> ax.plot(X, Y)
ax[0].plot(filas_messi["año"], # X
        filas_messi["goles"], # Y
        marker = 'o', # circulo
        color = 'b', # azul
        linestyle = '--', # dashed
        label = 'Lionel Messi') # etiqueta

ax[0].plot(filas_messi["año"], # X
        filas_messi["goles_sin_penalti"], # Y
        marker='.', # punto
        color='b', # azul
        linestyle='dotted') # dotted

ax[1].plot(filas_cristiano["año"], # X
        filas_cristiano["goles"], # Y
        marker = 's', # cuadrado
        color = 'r', # rojo
        linestyle = '--', # dashed
        label = 'Cristiano Ronaldo') # etiqueta

ax[1].plot(filas_cristiano["año"], # X 
        filas_cristiano["goles_sin_penalti"], # Y
        marker='.', # punto
        color='r', # rojo
        linestyle='dotted') # dotted

ax[2].plot(filas_cavani["año"], # X
        filas_cavani["goles"], # Y
        marker='D', # diamante
        color='m', # magenta
        linestyle='--', # dashed
        label='Edinson Cavani') # etiqueta

ax[2].plot(filas_cavani["año"], # X 
        filas_cavani["goles_sin_penalti"], # Y
        marker='.', # punto
        color='m', # magenta
        linestyle='dotted') # dotted

# Titulo, etiquetas
ax[0].set_title('Goles y no penalti goles')

ax[0].set_ylabel('Lionel Messi')
ax[1].set_ylabel('Cristiano Ronaldo')
ax[2].set_ylabel('Edinson Cavani')

ax[2].set_xlabel('Años')

# Leyenda
ax[0].legend()
ax[1].legend()
ax[2].legend()

# Mostramos el grafico
plt.show()
````

## Grafico de barras

El gráfico de barras se puede crear con **ax.bar(x, y)**.

**NOTA**

Si no se trabajara con **suplots**, podemos omitir el uso de **fig, ax = plt.subplots()** y trabajar todo directamente con **plt** en vez de con **ax**. 

En general, **plt** funciona muy parecido a **ax**, solo hay algunos métodos que cambian.

Por ejemplo, se usa **plt.xlabel** en lugar de **ax.set_xlabel** para etiquetar el eje X.

En los gráficos de barras es común hacer **comparaciones entre categorías**. Por lo tanto, en vez de tener una **escala de números**, podemos hacer que las **etiquetas** de los ejes sean **strings**, y probablemente *queden sobrepuestas unas sobre otras*. 

Para evitar este problema podemos usar **plt.xticks()**.

- **plt.xticks(rotation = 90)** → Rotara 90 grados las etiquetas de los ejes.

Tambien, podemos *cambiar los estilos por defecto de matplotlib*, ya que es posible usar **plantillas prediseñadas** con **plt.style.use()**. 

Se debe declarar antes de crear tlas figuras, para que se aplique a todas. Si queremos regresar a los estilos por defecto, podemos usar **'default'** como argumento.

[Plantillas Prediseñadas de Matplotlib](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)

````python
# Obtenemos las tarjetas rojas de cada equipo
equipos_tarjetas_rojas = df.groupby('nombre_equipo')['tarjetas_rojas'].sum()

equipos_tarjetas_rojas
````

````python
# Ordenamos descendentemente y obtenemos los 10 primeros
top_equipos_tarjetas_rojas = equipos_tarjetas_rojas.sort_values(ascending = False).head(10)

top_equipos_tarjetas_rojas
````

Con **.index** podemos acceder a los nombres de los equipos

````python
top_equipos_tarjetas_rojas.index
````

````python
# Usamos una plantilla prediseñada
plt.style.use('ggplot')

# Creamos un grafico de barras
# plt.bar(x, y)
plt.bar(top_equipos_tarjetas_rojas.index, top_equipos_tarjetas_rojas)

# Rotamos las etiquetas del eje X
plt.xticks(rotation = 90)

# Titulo, etiquetas
plt.title('Comparacion tarjetas rojas')
plt.ylabel('Tarjetas rojas')

# Mostramos el grafico
plt.show()
````

Si queremos que las barras estén horizontales en vez de verticales podemos usar **plt.barh()** en lugar de **plt.bar()**.

````python
# Usamos una plantilla prediseñada
plt.style.use('ggplot')

# Creamos un grafico de barras
# plt.bar(x, y)
plt.barh(top_equipos_tarjetas_rojas.index, top_equipos_tarjetas_rojas)

# Rotamos las etiquetas del eje X
plt.xticks(rotation = 90)

# Titulo, etiquetas
plt.title('Comparacion tarjetas rojas')
plt.ylabel('Tarjetas rojas')

# Mostramos el grafico
plt.show()
````

También podemos crear un **gráfico de barras apilado** para añadir más información. 

Lo único que tenemos que hacer es volver a llamar al método **.bar()** con una nueva variable y añadirle el parámetro **bottom = data_previa**. 

Y si queremos añadir aun más información, tendremos que sumar toda la información previa: 

**bottom = data_previa_1 + data_previa_2**.

Adenas, podemos usar **label** y **color** como argumentos de **.bar()** para etiquetar qué representa cada color.

````python
# Creamos una columna que sea la suma de goles y asistencias
df['goles_asistencias'] = df["goles"]+ df["asistencias"]

df['goles_asistencias']
````

````python
# Agrupamos por jugador y sumamos cada columna
agrupacion = df.groupby('nombre_jugador')[['goles', 'asistencias', 'goles_asistencias']].sum()

agrupacion
````

````python
# Ordenamos respecto a goles_asistencias y obtenemos los 10 primeros
top_jugadores = agrupacion.sort_values('goles_asistencias', ascending = False).head(10)

top_jugadores
````

````python
# Obtenemos los nombres de cada jugador
top_jugadores.index
````

````python
# Creamos los grafico de barras
plt.bar(top_jugadores.index, top_jugadores["goles"], label='Goles', color = "#55F25C")
plt.bar(top_jugadores.index, top_jugadores["asistencias"], label='Asistencias', bottom = top_jugadores["goles"], color = "#58D0DB")

# Rotamos las etiquetas del eje X
plt.xticks(rotation = 90)

# Titulo, etiquetas
plt.title('Goles y Asistencias Jugadores Top')
plt.ylabel('Goles y Asistencias')

# Leyenda
plt.legend()

# Mostrar grafico
plt.show()
````

## Histograma

Para graficar un **histograma** se usa **plt.hist(X)** y como argumento se pasa la columna que se queire analizar. 

Se puede personalizar el gráfico con los siguientes argumentos:

- **bins = cantidad** → Indica cuántas barras serán mostradas en el histograma. Por defecto son 10. A este argumento también le puedes pasar una lista para establecer un límite en el rango de las X en las que quieras que se muestren.
- **alpha** → Si queremos colocar 2 histogramas que se sobrepongan entre sí, podemos usar el argumento alpha que si se le pasa un valor entre 0 y 1 cambia la transparencia del color.
- **histtype='tipo'** → Este argumento hace que cambie la forma que se representa el histograma, por ejemplo líneas en vez de rectángulos sólidos.
  - [Tipos de histograma](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html).
- **density = True** → Si se usa, el gráfico se trasformará a **porcentajes** para que el *área total de los rectángulos sea de 1 (100%)*. Resulta útil para entender el peso de los valores más que su cantidad en sí misma.

````python
# Filtramos la data por año = 2018
data_2018 = df[df["año"] == 2018]

data_2018
````

````python
# Agrupamos por equipo y sumamos los goles
goles_equipo_2018 = data_2018.groupby('nombre_equipo')['goles'].sum()

goles_equipo_2018
````

````python
# Creamos una figura de tamaño 7 x 6 pulgadas (ancho, alto)
plt.figure(figsize=[7, 6])

# Creamos el histograma indicando que muestre 10 barras
plt.hist(goles_equipo_2018, bins = 10)

# Titulo, etiquetas
plt.title('Goles totales por equipo en 2018')
plt.xlabel('Cantidad de goles')
plt.ylabel('Número de equipos con X cantidad de goles')

# Mostrar grafico
plt.show()
````
